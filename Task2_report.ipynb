Отчет по второму заданию "Parallel sort"

Нужно сставить графики зависимости T(P), S(P), E(P) + на T(P) отобразить время библиотечной qsort. 

количество эл-тов в массиве и в чанке фиксированно

T(P)- время работы задачи на P потоках

S(P) = T(1) / T(P) - ускорение на P потоках

E(P) = S(P) / p - эффективность

```python
#Пусть n = 9.000.000, m = 90.000

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

results = [ [] for  i in range(5)]
i=0
j=0
f = open('stats.txt', 'r')
for line in f:
    results[i].append(line)
    j += 1
    if (j == 9):
        j = 0
        i += 1
        
time_on_N_9M_mysort = []
time_on_N_9M_qsort = []
P_with_9M = []
#забыл при первом выполнии прогнать на одном потоке
time_on_N_9M_mysort.append (2.017746)
time_on_N_9M_qsort.append(1.743863)
P_with_9M.append(1)

for i in range(0,4):
    time_on_N_9M_mysort.append(float(((results[i][8]).split(' '))[0]))
    time_on_N_9M_qsort.append(float(((results[i][8]).split(' '))[1]))
    P_with_9M.append(float(((results[i][8]).split(' '))[4]))
```

```python
   
plt.figure(figsize=(12, 6))

plt.plot(P_with_9M, time_on_N_9M_mysort, label="parallelSort")
plt.plot(P_with_9M, time_on_N_9M_qsort, label="qsort")
plt.title(r"$T(P)$", fontsize=14)
plt.xlabel("#threads")
plt.ylabel("seconds")
plt.xlim(0)
plt.ylim(0)
plt.grid("on")
plt.legend()
plt.show()
```

```python
S = (time_on_N_9M_mysort[0]) / np.array(time_on_N_9M_mysort) 

plt.figure(figsize=(12, 6))

plt.plot(P_with_9M, S, label="parallelSort, n=9*10^6, m = 9*10^4")
plt.title(r"$S(P)$", fontsize=14)
plt.xlabel("#threads")
plt.ylabel("acceleration")
plt.grid("on")
plt.xlim(0)
plt.ylim(0)
plt.legend()
plt.show()

```

```python
E = S / P_with_9M

plt.figure(figsize=(12, 6))

plt.plot(P_with_9M, E, label="parallelSort, n=9*10^6, m = 9*10^4")
plt.title(r"Efficiency from the number of threads", fontsize=14)
plt.xlabel("#threads")
plt.grid("on")
plt.xlim(0)
plt.ylim(0)
plt.legend()
plt.show()

```
