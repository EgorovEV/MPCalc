Решение задачи TSP путем применения генетического алгоритма.

```python
#Число шагов = 15
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


results = [[] for i in range(0,4)]
f = open('inter_info.txt', 'r')
i = 0
len_measurement = []
len = 0
for line in f:
    if (line == "\n"):
        len_measurement.append(len)
        len = 0
        i+=1
    else:
        results[i].append(line)
        len+=1
    ++len;
        
results_cost_min = [[] for i in range(0,3)]
results_cost_max = [[] for i in range(0,3)]
results_cost_avg = [[] for i in range(0,3)]
results_iteration = [[] for i in range(0,3)]

i = 0
k = 0
for x in range(0,3):
    for res in results[x]:
        if (i == len_measurement[k]):
            k += 1
            i = 0
        if (res == "\n"):
            break
        results_cost_min[k].append(int(res.split(' ')[0]))
        results_cost_max[k].append(int(res.split(' ')[1]))
        results_cost_avg[k].append(int(res.split(' ')[2]))
        results_iteration[k].append(i)
        i+=1
                    
population = [1000, 2000, 10000]
            
for i in range(0,3):
    plt.figure(figsize=(12, 6))
    plt.plot(results_iteration[i], results_cost_min[i], label="MIN")
    plt.plot(results_iteration[i], results_cost_max[i], label="MAX")
    plt.plot(results_iteration[i], results_cost_avg[i], label="AVG")
    s = "Population =" + str(population[i])
    plt.title(s, fontsize=14)
    plt.xlabel("iteration")
    plt.ylabel("minimum weight")
    plt.legend()
    plt.grid()
    plt.show()
```
