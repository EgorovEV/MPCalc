# ОТЧЕТ ПО ЗАДАЧЕ_4.
Решение задачи TSP путем применения генетического алгоритма.

Наблюдение: При выставалении числа шагов, за которые минимальный результат не должен измениться ~~ >20, 
точность значительно увеличивается (ровно как и время работы) Пример:

```python
#Число шагов = 25
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

results = []
f = open('inter_info.txt', 'r')
for line in f:
    results.append(line)

        
results_cost_min = []
results_cost_max = []
results_cost_avg = []
results_iteration = []

i = 0
for res in results:
    results_cost_min.append(int(res.split(' ')[0]))
    results_cost_max.append(int(res.split(' ')[1]))
    results_cost_avg.append(int(res.split(' ')[2]))
    results_iteration.append(i)
    i+=1
                    
    
plt.figure(figsize=(12, 6))
plt.plot(results_iteration, results_cost_min, label="MIN")
plt.plot(results_iteration, results_cost_max, label="MAX")
plt.plot(results_iteration, results_cost_avg, label="AVG")
plt.title(r"$population = 1000$", fontsize=14)
plt.xlabel("iteration")
plt.ylabel("minimum weight")
plt.legend()
plt.show()
#Случайно вывел 2 графика.
```
